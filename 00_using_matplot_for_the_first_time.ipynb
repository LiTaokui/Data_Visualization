```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3], [1,4,9])
plt.xlabel("x")
plt.ylabel("y")
plt.title("our first line")
plt.show()
```

```python
plt.plot([1,2,3], [1,4,9])
plt.plot([1,2,3], [10,20,30])
plt.xlabel("x")
plt.ylabel("y")
plt.title("our second line")
plt.legend(["line_1", "line_2"])
plt.show()
```
