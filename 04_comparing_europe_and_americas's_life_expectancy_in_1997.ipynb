```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/sample_data/countries.csv")
data
```

```python
set(data.continent)
```

```python
data_1997 = data[data.year == 1997]
data_1997
```

```python
europe_1997 = data_1997[data_1997.continent == "Europe"]
europe_1997
```

```python
america_1997 = data_1997[data_1997.continent == "Americas"]
america_1997
```

```python
len(europe_1997)
```

```python
len(america_1997)
```

```python
print("mean lifeExpectancy in europe")
print(europe_1997.lifeExpectancy.mean())

print("mean lifeExpectancy in america")
print(america_1997.lifeExpectancy.mean())

print("median lifeExpectancy in europe")
print(europe_1997.lifeExpectancy.median())

print("median lifeExpectancy in america")
print(america_1997.lifeExpectancy.median())
```

```python
plt.hist(europe_1997.lifeExpectancy, 10, edgecolor="black")
```

```python
plt.subplot(2,1,1)
plt.hist(europe_1997.lifeExpectancy, 20, range=(50,85), edgecolor="black")
plt.ylabel("europe")
plt.title("distribution of lifeExpectancy")

plt.subplot(2,1,2)
plt.hist(america_1997.lifeExpectancy, 20, range=(50,85), edgecolor="black")
plt.ylabel("america")
```
