```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/sample_data/countries.csv")
data
```

```python
data_2007 = data[data.year == 2007]
data_2007
```

```python
gdp_2007 = data_2007.population * data_2007.gdpPerCapita
gdp_2007
```

```python
top10 = data_2007.sort_values("population", ascending=False).head(10)
```

```python
x = range(10)
plt.bar(x, top10.gdpPerCapita * top10.population / 10**9)
plt.xticks(x, top10.country, rotation="vertical")
```

```python
plt.subplot(2, 1, 1)
x = range(10)
plt.bar(x, top10.population / 10**6)
plt.xticks([],[])
plt.title("10 most populus countries")
plt.legend(["population in millions"])

plt.subplot(2, 1, 2)
plt.bar(x, top10.gdpPerCapita * top10.population / 10**9)
plt.xticks(x, top10.country, rotation="vertical")
plt.legend(["GDP in billions"])
```
