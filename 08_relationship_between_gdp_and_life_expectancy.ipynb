```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("/content/sample_data/countries.csv")
data
```

```python
data_2007 = data[data.year == 2007]
data_2007.head()
```

```python
gdp_2007 = data_2007.gdpPerCapita * data_2007.population
gdp_2007
```

```python
plt.scatter(gdp_2007 / 10 ** 9, data_2007.lifeExpectancy )
plt.title("gdp and lifeExpectancy in 2007")
plt.xlabel("gdp per billion")
plt.ylabel("lifeExpectancy")
```

```python
(gdp_2007 / 10**9).corr(data_2007.lifeExpectancy)
```

```python
plt.scatter(np.log10(gdp_2007 / 10**9), data_2007.lifeExpectancy, 5)
```

```python
np.log10(gdp_2007 / 10**9).corr(data_2007.lifeExpectancy)
```

```python
year_sorted = sorted(set(data.year))
```

```python
for given_year in year_sorted:
  data_year = data[data.year == given_year]
  plt.scatter((data_year.population * data_year.gdpPerCapita / 10**9), data_year.lifeExpectancy, 5)
  plt.title(given_year)
  plt.xlim(0,1500)
  plt.ylim(25,85)
  plt.xlabel("gdp per billion")
  plt.ylabel("lifeExpectancy")
  plt.show()
  plt.savefig(str(given_year))
```

```python
for given_year in year_sorted:
  data_year = data[data.year == given_year]
  plt.scatter(np.log10(data_year.population * data_year.gdpPerCapita / 10**9), data_year.lifeExpectancy, 5)
  plt.title(given_year)
  plt.xlim(0,5)
  plt.ylim(25,85)
  plt.xlabel("gdp per billion")
  plt.ylabel("lifeExpectancy")
  plt.show()
  plt.savefig("log" + str(given_year))
```

```python
np.log10(data_year.population * data_year.gdpPerCapita / 10**9).corr(data_year.lifeExpectancy)
```

```python
(data_year.population * data_year.gdpPerCapita / 10**9).corr(data_year.lifeExpectancy)
```
